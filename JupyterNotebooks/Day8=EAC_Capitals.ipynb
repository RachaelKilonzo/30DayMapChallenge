```python
import os
import osmnx as ox
from IPython.display import Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

# Set the working directory
path = "C:/Users/PC/Desktop/Portfolio/MapChallenge/Day8/"
os.chdir(path)

# Configure OSMnx settings
ox.settings.log_file = True
ox.settings.log_console = True
ox.settings.use_cache = True
```

```python
# Configure image display settings
img_folder = "C:/Users/PC/Desktop/Portfolio/MapChallenge/Day8/images"
extension = 'png'
size = 350
dpi = 900

# Configure street width pixels
street_widths = {'footway': 0.5,
                 'steps': 0.5,
                 'pedestrian': 0.5,
                 'path': 0.5,
                 'track': 0.75,
                 'service': 1.5,
                 'residential': 2.5,
                 'primary': 2.5,
                 'motorway': 3
                }

# Function to create and save the diagram
def create_plot(point, place):
    G = ox.graph_from_point(point, dist=10000, network_type='all')
    fig, ax = ox.plot_figure_ground(G, street_widths=street_widths, dpi=dpi)
    filepath = f"{img_folder}/{place}.{extension}"
    fig.savefig(filepath, dpi=dpi)
    return Image(filepath, height=size, width=size)

# Create and display the diagrams
create_plot((-1.2864, 36.8213), "Nairobi")
create_plot((0.3270, 32.5858), "Kampala")
create_plot((-6.18130, 35.74798), "Dar Es Salaam")
create_plot((-1.9577, 30.0860), "Kigali")
```

```python
# List of file paths
file_paths = ["C:/Users/PC/Desktop/Portfolio/MapChallenge/Day8/images/Nairobi.png",
              "C:/Users/PC/Desktop/Portfolio/MapChallenge/Day8/images/Kampala.png",
              "C:/Users/PC/Desktop/Portfolio/MapChallenge/Day8/images/Dar Es Salaam.png",
              "C:/Users/PC/Desktop/Portfolio/MapChallenge/Day8/images/Kigali.png"]

# Load images and store their arrays in a list
images = [mpimg.imread(file) for file in file_paths]

# Concatenate images horizontally
combined_image = np.concatenate(images, axis=1)

# Save the combined image
output_file = "C:/Users/PC/Desktop/Portfolio/MapChallenge/Day8/images/combined_image.png"
plt.imsave(output_file, combined_image)

# Display the combined image (optional)
plt.imshow(combined_image)
plt.axis('off')
plt.show()
```
